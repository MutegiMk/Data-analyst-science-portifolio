````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

````

````python
# loading the dataset
credit_card_data = pd.read_csv('/content/creditcard.csv') 
````

````python
# printint 5 rows
credit_card_data.head()
````

````python
# dataset information
credit_card_data.info()
````

````python
# Descriptive statistics
credit_card_data.describe()
````

````python
# values counts fronthe two transactions
credit_card_data['Class'].value_counts(), print('data is highly unbalanced')
````

0 -> Normal Transaction

1 -> fraudulent Transaction 

````python
# seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0.0]
fraud = credit_card_data[credit_card_data.Class == 1.0]
````

````python
print(legit.shape)
print(fraud.shape)
````

````python
# Statistical measures of the data
legit.Amount.describe(), fraud.Amount.describe()
````

````python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
````

Under-Sampling

Building a sample dataset containing similar distribution of legit and fraud transactions

fraud transaction = 239

````python
legit_sample = legit.sample(n=239)
````

joining the two dataframes

````python
balanced_dataset = pd.concat([legit_sample, fraud], axis=0)
````

````python
balanced_dataset.head()
````

````python
balanced_dataset['Class'].value_counts()
````

````python
balanced_dataset.groupby('Class').mean()
````

Splitting the data into features and Targets


````python
X = balanced_dataset.drop(columns='Class', axis=1)
y= balanced_dataset["Class"]
````

split the data into Training data and test data

````python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=2 )
````

````python
print(X.shape, X_train.shape, X_test.shape)
````

Model Training

````python
model =LogisticRegression()
````

````python
# Trainng the model
model.fit(X_train,y_train)
````

Model Evaluation


````python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_train_prediction,y_train)
````

````python
print('Accuracy on Training data : ', training_accuracy )
````

````python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
````

````python
print('Accuracy on Training data : ', test_data_accuracy )
````
